```python
import numpy as np
import pandas as pd
from collections import Counter
```

```python
rng = np.random.default_rng(42)

impostors = {1}
votes = []
game_id = 1
round_id = 1
match_id = 0
votes = []
game_results = []
votes_for_round = []
impostors_wins = False 
R_values = list(range(1,6))
P_values = list(range(3,10))

for p in P_values:
    players = []
    players = list(range(p))
    players_alive = players.copy()
    for r in R_values:
        for m in  range(0, 1000):
            match_id += 1
            round_id = 1
            players_alive = players.copy()
            for i in range(r):
                impostors_wins = False
                game_id = 1
                impostors ={rng.choice(players)}
                players_alive = players.copy()
                while any(imp in players_alive for imp in impostors) and not impostors_wins:
                    votes_for_round = []
                    for voter in players_alive:
                        # no se vota a sí mismo
                        possible_targets = [p for p in players_alive if p != voter]

                        target = rng.choice(possible_targets)

                        votes.append({
                            "P": p,
                            "R": r,
                            "match": match_id,
                            "game": game_id,
                            "round": round_id,
                            "voter": voter,
                            "target": target,
                            "voter_role": "impostor" if voter in impostors else "civil",
                            "target_role": "impostor" if target in impostors else "civil"
                        })

                        votes_for_round.append(
                             target
                        )
                    count_votes = Counter(votes_for_round)
                    (max_voted, quantity_of_votes_for_most_voted) = count_votes.most_common(1)[0]
                    most_voted = []
                    most_voted = [ player for player, votes in count_votes.items()
                                if votes == quantity_of_votes_for_most_voted]                                                                


                    if len(most_voted) > 1:
                        #integer = rng.integers(0, len(most_voted))
                        #eliminated = most_voted.index[integer]
                        eliminated = rng.choice(most_voted)
                        players_alive.remove(eliminated)
                    else:
                        eliminated = most_voted[0]
                        players_alive.remove(eliminated)

                    alive_set = set(players_alive)
                    impostors_wins = len(alive_set & impostors) >= len(alive_set - impostors)

                    game_results.append({
                        "P": p,
                        "R": r,
                        "match_id": match_id,
                        "round": round_id,
                        "game_id": game_id,
                        "eliminated": eliminated,
                        "impostors_wins": impostors_wins
                        })
                    game_id += 1
                round_id += 1

                    
                

            
        
        
        

```

```python
votes_df = pd.DataFrame(votes)
results_df = pd.DataFrame(game_results)
```

```python
acc_by_PR = (
    votes_df
    .groupby(["P","R"])["correct_vote"]
    .mean()
    .reset_index(name="accuracy")
)

acc_by_PR.head(10)
```

```python
score_df = (
    votes_df
    .groupby(["P", "R", "match", "voter"])["correct_vote"]
    .sum()
    .reset_index(name="points")
)

```

```python
score_df
```

```python
score_stats = (
    score_df
    .groupby(["P", "R"])["points"]
    .agg(mean_points="mean", std_points="std")
    .reset_index()
)
```

```python
score_stats["cv"] = score_stats["std_points"] / score_stats["mean_points"]
```

```python
score_stats
```

```python
score_stats[score_stats["P"] == 6].plot(
    x="R",
    y="cv",
    marker="o",
    title="Coeficiente de variación vs R (P=6)"
)
```
