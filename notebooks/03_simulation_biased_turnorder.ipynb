```python
import sys
sys.executable
import numpy as np
import pandas as pd
from collections import Counter
```

```python
rng = np.random.default_rng(42)

impostors = {1}
votes = []
game_id = 1
round_id = 1
match_id = 0
votes = []
game_results = []
votes_for_round = []
impostors_wins = False 
R_values = list(range(1,6))
P_values = list(range(3,10))
alpha = 2



for p in P_values:
    players = []
    players = list(range(p))
    players_alive = players.copy()
    last_speaker_id = players_alive[-1]
    for r in R_values:
        for m in  range(0, 1000):
            match_id += 1
            round_id = 1
            players_alive = players.copy()
            for i in range(r):
                impostors_wins = False
                game_id = 1
                impostors ={rng.choice(players)}
                players_alive = players.copy()
                for speakers_order_id, voter_value in enumerate(players_alive):
                        speakers_order = speakers_order_id
                        voter = voter_value
                while any(imp in players_alive for imp in impostors) and not impostors_wins:
                    votes_for_round = []
                    
                    for voter in players_alive: 
                        
                        weights = [] 
                        for new_voter in players_alive: 
                            if new_voter != voter:
                                if new_voter == players_alive[-1]:
                                    weights.append(1 + alpha)
                                    probs = [w/sum(weights) for w in weights]
                                else:
                                    weights.append(1)
                                    probs = [w/sum(weights) for w in weights]
                        
                        last_speaker = players_alive[-1]
                        
                        # no se vota a sÃ­ mismo
                        possible_targets = [p for p in players_alive if p != voter]

                        is_last_speaker = players_alive[-1]
                        

                        target = rng.choice(possible_targets, p=probs)
                        votes.append({
                            "P": p,
                            "R": r,
                            "match": match_id,
                            "game": game_id,
                            "round": round_id,
                            "voter": voter,
                            "target": target,
                            "voter_role": "impostor" if voter in impostors else "civil",
                            "target_role": "impostor" if target in impostors else "civil",
                            "turn_of_speak": speakers_order,
                            "is_the_last_turn_of_speaker" : (last_speaker == voter),
                            "vote_is_correct": target == "impostor" in impostors
                        })

                        votes_for_round.append(
                             target
                        )
                    count_votes = Counter(votes_for_round)
                    (max_voted, quantity_of_votes_for_most_voted) = count_votes.most_common(1)[0]
                    most_voted = []
                    most_voted = [ player for player, votes in count_votes.items()
                                if votes == quantity_of_votes_for_most_voted]                                                                


                    if len(most_voted) > 1:
                        #integer = rng.integers(0, len(most_voted))
                        #eliminated = most_voted.index[integer]
                        eliminated = rng.choice(most_voted)
                        players_alive.remove(eliminated)
                    else:
                        eliminated = most_voted[0]
                        players_alive.remove(eliminated)

                    alive_set = set(players_alive)
                    impostors_wins = len(alive_set & impostors) >= len(alive_set - impostors)

                    game_results.append({
                        "P": p,
                        "R": r,
                        "match_id": match_id,
                        "round": round_id,
                        "game_id": game_id,
                        "eliminated": eliminated,
                        "impostors_wins": impostors_wins
                        })
                    game_id += 1
                round_id += 1

                    
                

            
        
        
        

```
